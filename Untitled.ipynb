```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 
           'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 
           'hours-per-week', 'native-country', 'income']
data = pd.read_csv(url, names=columns, sep=',\s+', engine='python')

# Step 2: Preprocess the dataset
# Strip spaces from columns
data['income'] = data['income'].str.strip()
data['race'] = data['race'].str.strip()
data['sex'] = data['sex'].str.strip()

# Step 3: Calculate bias metrics
# Group by gender and race, and calculate the proportion of people earning >50K
grouped_data = data.groupby(['race', 'sex', 'income']).size().unstack(fill_value=0)
grouped_data['Total'] = grouped_data['<=50K'] + grouped_data['>50K']
grouped_data['Proportion >50K'] = grouped_data['>50K'] / grouped_data['Total']

# Step 4: Visualize the bias using a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='sex', y='Proportion >50K', hue='race', data=grouped_data.reset_index())
plt.title('Income Distribution by Gender and Race')
plt.ylabel('Proportion Earning >50K')
plt.xlabel('Gender')
plt.show()

# Step 5: Print summary table
summary_table = grouped_data[['Proportion >50K']].reset_index()
print(summary_table)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from fairlearn.metrics import MetricFrame, accuracy_score
from fairlearn.metrics import demographic_parity_difference, equalized_odds_difference
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

# Step 1: Load the dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 
           'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 
           'hours-per-week', 'native-country', 'income']
data = pd.read_csv(url, names=columns, sep=',\s+', engine='python')

# Step 2: Preprocess the data
# Strip spaces from column values in case there are any
data['income'] = data['income'].str.strip()
data['sex'] = data['sex'].str.strip()
data['race'] = data['race'].str.strip()

# Encode the categorical data
label_enc = LabelEncoder()
data['income'] = label_enc.fit_transform(data['income'])  # 0 for <=50K, 1 for >50K

# Selecting relevant features
X = data[['age', 'education-num', 'hours-per-week']]
y = data['income']

# Step 3: Train a simple model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Step 4: Define sensitive features for fairness check
# Ensure we can access the sensitive columns
sensitive_features = pd.DataFrame()  # Create a DataFrame to store sensitive features
sensitive_features['sex'] = data.loc[X_test.index, 'sex']  # Ensure alignment
sensitive_features['race'] = data.loc[X_test.index, 'race']

# Step 5: Calculate fairness metrics
dem_parity_gender = demographic_parity_difference(y_true=y_test, y_pred=y_pred, sensitive_features=sensitive_features['sex'])
equal_opp_gender = equalized_odds_difference(y_true=y_test, y_pred=y_pred, sensitive_features=sensitive_features['sex'])

print(f"Demographic Parity Difference by Gender: {dem_parity_gender}")
print(f"Equalized Odds Difference by Gender: {equal_opp_gender}")

# Step 6: Use MetricFrame for general metrics like accuracy
metric_frame = MetricFrame(metrics=accuracy_score,
                           y_true=y_test,
                           y_pred=y_pred,
                           sensitive_features=sensitive_features['sex'])

# Display accuracy by group
print("\nAccuracy by Gender:")
print(metric_frame.by_group)

# Step 7: Visualize the accuracy breakdown
metric_frame.by_group.plot(kind='bar', title="Accuracy by Gender")
plt.ylabel("Accuracy")
plt.show()
```
